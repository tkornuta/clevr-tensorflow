```python
#Python
import tensorflow as tf

sess = tf.InteractiveSession()

# Structure the graph according to the 3-stage pattern.
# 1. Inference
with tf.name_scope("inference"):
  # Create a linear neuron model.
  W = tf.Variable([.3], dtype=tf.float32, name= "W")
  b = tf.Variable([-.3], dtype=tf.float32, name="b")
  x = tf.placeholder(tf.float32, name = "x")
  linear_model = W * x + b


# 2. MSE loss function.
with tf.name_scope("loss"):
  y = tf.placeholder(tf.float32, name="y")
  squared_deltas = tf.square(linear_model - y)
  loss = tf.reduce_sum(squared_deltas)
  loss_summary = tf.summary.scalar('loss', loss)

# 3. Training
# Add optimization function - simple SGD.
with tf.name_scope("training"):
  optimizer = tf.train.GradientDescentOptimizer(0.01)
  train = optimizer.minimize(loss)

# Training data.
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]

# Initialize values (to "incorrect" defaults).
init = tf.global_variables_initializer()
sess.run(init)

# Merge all the summaries...
#merged = tf.summary.merge_all()
# ... and write them out to given dir.
writer = tf.summary.FileWriter('/tmp/TF/simple_model/', sess.graph)

# Training.
print("Training the model. Please wait...")
for i in range(1000):
  sess.run(train, {x:x_train, y:y_train})
  if i % 1 == 0:  # Record summaries
    print("i =", i, "loss =",sess.run(loss, {x:x_train, y:y_train}))
    summary = sess.run(loss_summary, feed_dict={x:x_train, y:y_train})
    writer.add_summary(summary, global_step=i)

# Evaluate the training accuracy.
curr_W, curr_b, curr_loss = sess.run([W, b, loss], {x:x_train, y:y_train})
print("W: %s b: %s \nloss: %s"%(curr_W, curr_b, curr_loss))

writer.flush()
sess.close()
```
