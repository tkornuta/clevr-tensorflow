```python
import numpy as np
import matplotlib.pyplot as plt

plt.subplots_adjust(hspace=0.4)
t = np.arange(-10.0, 10.0, 0.01)
t_pos = np.arange(-0.7, 10.0, 0.01)

plt.subplot(221)
plt.plot(t, np.exp(t))
plt.title('e^t')
plt.grid(True)

plt.subplot(222)
plt.plot(t_pos, np.log1p(t_pos))
plt.title('log1p(t_pos)')
plt.grid(True)

plt.subplot(223)
plt.plot(t, np.log1p(np.exp(t)))
plt.grid(True)
plt.title('log1p(e^t)')

ax = plt.subplot(224)
plt.plot(t, 1 + np.log1p(np.exp(t)))
plt.grid(True)
plt.title('1 + log1p(e^t)')

plt.show()
```
